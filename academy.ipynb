```python
import pandas as pd
import numpy as np
import pyensae

#------------- Fonctions -----------------------
## nombre de valeurs null
def compteNull (envoiData):
    #print(" envoiData :", envoiData)
    a = envoiData.T['null'].sum()
    b = envoiData.T['count'].sum()
    print("Nbre valeurs 'null' : {:.0f}".format(a))
    print("Nbre valeurs non 'null' : {:.0f}".format(b))
    print("Nbre total cases : {:.0f}".format(a+b))
    print("% total valeurs 'null' : {:.1f}%".format(a*100/(a+b)))

## informations sur le data frame

def infoFrame (df):
    nb_li = df.index.size
    nb_col = df.columns.size
    tot = nb_li*nb_col    
    infos = pd.DataFrame(df.dtypes).T.rename(index={0:'Type'}) 
    infos = infos.append(pd.DataFrame(df.isna().sum()).T.rename(index={0:'null'}))
    return infos
#------------------------------------------

# Lire les fichier csv dans un dataframe pandas avec suppression des colonne null

dataFile = pd.read_csv("C:\\environnementPython\\env01\\share\\Academy\\EdStatsData.csv").dropna(how='all', axis='columns')
countryFile = pd.read_csv("C:\\environnementPython\\env01\\share\\Academy\\EdStatsCountry.csv").dropna(how='all', axis='columns')
cnt_serFile = pd.read_csv("C:\\environnementPython\\env01\\share\\Academy\\EdStatsCountry-Series.csv").dropna(how='all', axis='columns')
seriesFile = pd.read_csv("C:\\environnementPython\\env01\\share\\Academy\\EdStatsSeries.csv").dropna(how='all', axis='columns')
footnoteFile = pd.read_csv("C:\\environnementPython\\env01\\share\\Academy\\EdStatsFootNote.csv").dropna(how='all', axis='columns')




```
